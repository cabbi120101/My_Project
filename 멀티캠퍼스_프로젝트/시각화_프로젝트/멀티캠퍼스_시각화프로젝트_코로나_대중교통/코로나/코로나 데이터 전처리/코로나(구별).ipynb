```python
data_file = '/content/data/코로나(서울구별데이터처리).csv'
df = pd.read_csv(data_file)
df.head()
```

```python
df.info()
```

```python
#지역 확인
df['지역'].unique()
```

```python
# 필요없는 타시도, 기타 추출
df = df[df['지역']!='타시도' ].copy()
df = df[df['지역']!='기타' ].copy()
df['지역'].unique()
```

```python
# 타입확인
addr = df['지역']
type(addr)
```

```python
# 공백이 있는 것들 없는 것으로 교체
df = df.replace({' 중랑구':'중랑구','강남구 ':'강남구','송파구 ' : '송파구','서초구 ':'서초구','마포구 ':'마포구','강서구 ':'강서구','서대문구 ':'서대문구'})
```

```python
# 공백있는 값들이 없어짐을 확인
df['지역'].value_counts()
```

```python
# Serise으로 리턴
print(type(df['확진일'].str.split('-')))
print(df['확진일'].str.split('-'))
#print(df['variable'].str.split('.')[1][0])
#print(df['variable'].str.split('.')[1][1])

# DataFrame으로 리턴
print(type(df['확진일'].str.split('-', expand=True)))
df['확진일'].str.split('-', expand=True).head()
```

```python
#확진일을 연도, 월, 성별로 분류
df['연도'] = df['확진일'].str.split('-', expand=True)[0]
df['월'] = df['확진일'].str.split('-', expand=True)[1]
df['일'] = df['확진일'].str.split('-', expand=True)[2]
#print(df['확진일'].str.split('-'))
```
