```python
from urllib.request import urlopen, Request
from urllib.parse import urlencode, unquote, quote_plus
import requests
import json
import pandas as pd
import xmltodict # 결과가 xml 형식으로 반환된다. 이것을 dict 로 바꿔주는 라이브러리다
```

```python
key = '5975486f6d636162373245774d6b79'

num_of_rows = 10 # 한페이지에 포함된 결과 수 
page_no = 1 #페이지 번호
url = 'http://openapi.seoul.go.kr:8088/sample/xml/Corona19Status/1/5/'
start = 0
end = 99999999
type = json
service = 'Corona19Status'

req_parameter = '?' + urlencode({
    'KEY':  key,
    'numOfRows' : num_of_rows,
    'pageNo' : page_no,
    'START_INDEX' : start,
    'END_INDEX' : end,
    'SERVICE' : service,
    'TYPE': type
})

r = requests.get(url + req_parameter)
response_content = json.loads(json.dumps(xmltodict.parse(r.text)))
response_content['Corona19Status']['row']
```

```python
row = response_content['Corona19Status']['row']
df = pd.DataFrame(row)
df.to_csv('/content/drive/코로나(서울시구별).csv', index = False)
```
